# Let's take a look at some of the most basic Keras functionality

## Classification

Let's build a Neural Net that can classify on the Iris set. This is overkill to the 1e6'th degree... but let's just prove we can do it.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils

X = load_iris().data
y = load_iris().target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
ohe = LabelEncoder().fit(y_train)
y_train = ohe.transform(y_train)
y_test_t = ohe.transform(y_test)
y_train_t = np_utils.to_categorical(y_train)
y_test_t = np_utils.to_categorical(y_test_t)
```

```python
y_train[:5]
```

```python
# Create your first MLP in Keras
from keras.models import Sequential
from keras.layers import Dense
import numpy

model = Sequential()
model.add(Dense(12, input_dim=4, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(3, activation='softmax'))

# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Fit the model
model.fit(X_train, y_train_t, epochs=150, batch_size=10)
```

```python
# evaluate the model
scores = model.evaluate(X_train, y_train_t)
scores
```

```python
scores = model.evaluate(X_test, y_test_t)
scores
```

```python
preds = model.predict(X_test)
preds_class = model.predict_classes(X_test)
```

```python
print(preds_class[0],preds[0],y_test[0])
```

## Now let's do some regression

We'll just work with a sine wave. And let's look how different levels of training work out for us.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(0.0, 1, 0.01).reshape(-1, 1)
y = np.sin(2 * np.pi * x).ravel()
test_x = np.arange(0.0, 1, 0.05).reshape(-1, 1)
true_y = np.sin(2 * np.pi * test_x).ravel()

plt.figure(dpi=150)
plt.plot(x,y, 'r', label="Truth")
```

```python
def run_model_and_plot(layer1_num=10, layer2_num=10, epochs=10, activ='tanh', return_model=False):
    model = Sequential()
    model.add(Dense(layer1_num, input_dim=1, activation=activ))
    model.add(Dense(layer2_num, activation=activ))
    model.add(Dense(1))

    # Compile model
    model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mean_squared_error'])

    # Fit the model
    model.fit(x, y, epochs=epochs, batch_size=2, verbose=0)
    preds = model.predict(x)

    plt.figure(dpi=150)
    plt.scatter(x,preds, c='k', label="Prediction");
    plt.plot(x,y, 'r', label="Truth")
    plt.legend(loc='upper right')
    if return_model:
        return model
```

```python
run_model_and_plot()
```

```python
run_model_and_plot(epochs=100)
```

```python
run_model_and_plot(epochs=250)
```

```python
model = run_model_and_plot(epochs=350)
```

```python
model = run_model_and_plot(return_model=True)
```
